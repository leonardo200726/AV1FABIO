## PARTE 3 ‚Äì Encapsulamento e Abstra√ß√£o

**Tarefas:**
- Tornar pelo menos 1 atributo da classe principal privado (`__atributo`).
- Criar m√©todos `get` e `set` para o atributo privado.
- Implementar um m√©todo que contenha l√≥gica interna (ex: c√°lculo, verifica√ß√£o).

**C√≥digo:**

```python
class Aluno_Leonardo:
    def __init__(self, nome, notas):
        self.__nome = nome  # Atributo privado
        self.__notas = notas
        
    def calcular_media_Leonardo(self):
        media = sum(self.__notas) / len(self.__notas)
        return round(media, 2)
    
    def exibir_dados_Leonardo(self):
        media = self.calcular_media_Leonardo()
        return f"Aluno: {self.__nome}, Notas: {self.__notas}, M√©dia: {media}"
    
    def adicionar_nota_Leonardo(self, nova_nota):
        self.__notas.append(nova_nota)
        
    def remover_nota_Leonardo(self, nota):
        if nota in self.__notas:
            self.__notas.remove(nota)
    
    # M√©todos get e set para o atributo privado __nome
    def get_nome_Leonardo(self):
        return self.__nome
    
    def set_nome_Leonardo(self, novo_nome):
        self.__nome = novo_nome
    
    # M√©todo com l√≥gica interna: verifica se o aluno passou ou reprovou
    def verificar_status_Leonardo(self):
        media = self.calcular_media_Leonardo()
        if media >= 7.0:
            return f"{self.__nome} est√° aprovado!"
        else:
            return f"{self.__nome} est√° reprovado!"
    

class Aluno_Graduacao_Leonardo(Aluno_Leonardo):
    def __init__(self, nome, notas, curso):
        super().__init__(nome, notas)
        self.curso = curso
        
    def exibir_dados_Graduacao_Leonardo(self):
        dados = super().exibir_dados_Leonardo()
        return f"{dados}, Curso: {self.curso}"
    
    def alterar_curso_Leonardo(self, novo_curso):
        self.curso = novo_curso
```

```python
import sqlite3
import json

class Aluno_Leonardo:
    def __init__(self, nome, notas):
        self.__nome_Leonardo = nome
        self.__notas_Leonardo = notas
        
    def calcular_media_Leonardo(self):
        media = sum(self.__notas_Leonardo) / len(self.__notas_Leonardo)
        return round(media, 2)
    
    def exibir_dados_Leonardo(self):
        media = self.calcular_media_Leonardo()
        return f"Aluno: {self.__nome_Leonardo}, Notas: {self.__notas_Leonardo}, M√©dia: {media}"
    
    def adicionar_nota_Leonardo(self, nova_nota):
        self.__notas_Leonardo.append(nova_nota)
        
    def remover_nota_Leonardo(self, nota):
        if nota in self.__notas_Leonardo:
            self.__notas_Leonardo.remove(nota)
    
    def get_nome_Leonardo(self):
        return self.__nome_Leonardo
    
    def set_nome_Leonardo(self, novo_nome):
        self.__nome_Leonardo = novo_nome
    
    def verificar_status_Leonardo(self):
        media = self.calcular_media_Leonardo()
        if media >= 7.0:
            return f"{self.__nome_Leonardo} est√° aprovado!"
        else:
            return f"{self.__nome_Leonardo} est√° reprovado!"
    
    def salvar_no_banco_Leonardo(self):
        connection = sqlite3.connect('alunos_Leonardo.db')
        cursor = connection.cursor()
        cursor.execute('''CREATE TABLE IF NOT EXISTS alunos_Leonardo (
                            id INTEGER PRIMARY KEY,
                            nome TEXT,
                            notas TEXT,
                            curso TEXT)''')
        
        notas_json = json.dumps(self.__notas_Leonardo)
        cursor.execute('INSERT INTO alunos_Leonardo (nome, notas) VALUES (?, ?)', 
                       (self.__nome_Leonardo, notas_json))
        
        connection.commit()
        connection.close()

class Aluno_Graduacao_Leonardo(Aluno_Leonardo):
    def __init__(self, nome, notas, curso):
        super().__init__(nome, notas)
        self.curso_Leonardo = curso
        
    def exibir_dados_Leonardo(self):
        dados = super().exibir_dados_Leonardo()
        return f"{dados}, Curso: {self.curso_Leonardo}"
    
    def alterar_curso_Leonardo(self, novo_curso):
        self.curso_Leonardo = novo_curso
    
    def salvar_no_banco_Leonardo(self):
        connection = sqlite3.connect('alunos_Leonardo.db')
        cursor = connection.cursor()
        cursor.execute('''CREATE TABLE IF NOT EXISTS alunos_Leonardo (
                            id INTEGER PRIMARY KEY,
                            nome TEXT,
                            notas TEXT,
                            curso TEXT)''')
        
        notas_json = json.dumps(self.__notas_Leonardo)
        cursor.execute('INSERT INTO alunos_Leonardo (nome, notas, curso) VALUES (?, ?, ?)', 
                       (self.get_nome_Leonardo(), notas_json, self.curso_Leonardo))
        
        connection.commit()
        connection.close()

    def consultar_aluno_Leonardo(self, nome):
        connection = sqlite3.connect('alunos_Leonardo.db')
        cursor = connection.cursor()
        cursor.execute('SELECT * FROM alunos_Leonardo WHERE nome = ?', (nome,))
        aluno = cursor.fetchone()
        connection.close()
        return aluno

    def excluir_aluno_Leonardo(self, nome):
        connection = sqlite3.connect('alunos_Leonardo.db')
        cursor = connection.cursor()
        cursor.execute('DELETE FROM alunos_Leonardo WHERE nome = ?', (nome,))
        connection.commit()
        connection.close()

# Testando as funcionalidades
aluno1_Leonardo = Aluno_Leonardo("Carlos", [7.5, 8.0, 9.0])
aluno1_Leonardo.salvar_no_banco_Leonardo()

aluno2_Graduacao_Leonardo = Aluno_Graduacao_Leonardo("Maria", [6.0, 7.0, 8.5], "Engenharia")
aluno2_Graduacao_Leonardo.salvar_no_banco_Leonardo()

# Consultar um aluno no banco
print(aluno2_Graduacao_Leonardo.consultar_aluno_Leonardo("Maria"))

# Excluir um aluno do banco
aluno2_Graduacao_Leonardo.excluir_aluno_Leonardo("Maria")

# Consultar ap√≥s exclus√£o
print(aluno2_Graduacao_Leonardo.consultar_aluno_Leonardo("Maria"))
```
